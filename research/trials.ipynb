```python
## FETCHING THE CATEGORICAL_COLUMN
Categorical_column = []

for x in DF.dtypes.index:
    if DF.dtypes[x] == 'object':
        Categorical_column.append(x)

Categorical_column
```

```python
## We are going to remove this columns becoz they are unique ids and not much of a help in model building
Categorical_column.remove('Item_Identifier')
Categorical_column.remove('Outlet_Identifier')
Categorical_column
```

```python
for col in Categorical_column:
    print(col)
    print(DF[col].value_counts())
    print()

## value counts of columns
```

```python
item_weight_mean = DF.pivot_table(values = "Item_Weight", index = 'Item_Identifier')
item_weight_mean

## Basically this are item weight mean values for different items identifier groups
```

```python
miss_bool = DF['Item_Weight'].isnull()
miss_bool
```

```python
import numpy as np
for i, item in enumerate(DF['Item_Identifier']):
    if miss_bool[i]:
        if item in item_weight_mean:
            DF['Item_Weight'][i] = item_weight_mean.loc[item]['Item_Weight']
        else:
            DF['Item_Weight'][i] = np.mean(DF['Item_Weight'])

## Fill the values of column item_weight with mean values 
```
